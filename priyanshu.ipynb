```python
import requests
from bs4 import BeautifulSoup
```

```python
location = []
title=[]
status=[]
size = []
price = []
pricepersqfeet = []
city = []
cityname = []

for i in range(1,50):
    URL="https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page={}&_=1664449405577".format(i)
    res = requests.get(URL)

    data = res.text
    bs_data = BeautifulSoup(data)
    b =bs_data.find_all("div", attrs = { "class": "locWrap"})
    for j in b:
        location.append(j.text)
    a=bs_data.find_all( "div", attrs = { "class":"title-line"})
    for j in a:
        title.append(j.text)
    c = bs_data.find_all("td",attrs = { "class": "val"})
    for j in c: 
        status.append(j.text)
    d = bs_data.find_all("td",attrs = {"class": "size"})
    for j in d:
        size.append(j.text)
        
    e = bs_data.find_all("td" ,attrs ={"class": "price" })
    for j in e:
        price.append(j.text)
      
    g = bs_data.find_all("td",attrs = {"class": "lbl rate"})
    for j in g :
        pricepersqfeet.append(j.text)
    h = bs_data.find_all("span",attrs = {"class":"cityName"})
    for j in h :
         cityname.append(j.text)
   
```
